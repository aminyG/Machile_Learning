```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Load dataset
iris = datasets.load_iris()
data = iris.data
```

```python
# Data normalization
data = data / data.max(axis=0)
```

```python
# Initialize SOM
map_size = (10, 10)
som = MiniSom(map_size[0], map_size[1], data.shape[1], sigma=0.5, learning_rate=0.5)
```

```python
# Initialize weights randomly
som.random_weights_init(data)
```

```python
# SOM Training
num_epochs = 100
som.train_random(data, num_epochs)
```

```python
# Visualization of SOM results
plt.figure(figsize=(8, 8))
for i, x in enumerate(data):
     w = som.winner(x) # Winner for sample x
     plt.text(w[0]+.5, w[1]+.5, str(i+1), color='k', fontdict={'weight': 'bold', 'size': 11} )
plt.pcolor(som.distance_map().T, cmap='bone_r', alpha=.2)
plt.colorbar()

plt.show()
```
